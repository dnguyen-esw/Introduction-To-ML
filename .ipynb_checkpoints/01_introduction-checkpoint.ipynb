# A First Application: Classifying Iris Species

<img src="iris.jpg" style="width:25%;height:25%">

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

## Keys của dataset

```python
print("Keys of iris_dataset: \n{}".format(iris_dataset.keys()))
```

## Value của key 'DESCR' là mô tả về bộ dataset

```python
print(iris_dataset['DESCR'][:1208] + "\n...")
```

## Value của key 'target_names' là một mảng chứa các chuỗi, là loại hoa muốn dự đoán => Đầu ra

```python
print("Target names: {}".format(iris_dataset['target_names']))
```

##  Value của key 'feature_names' là một list các chuỗi, mô tả các feature trong dataset

```python
print("Feature names: \n{}".format(iris_dataset['feature_names']))
```

## Value của key 'data' chứa giá trị của các 'feature_names' với tập dữ liệu 150 mẫu. Là 1 ma trận 150x4

```python
print("Type of data: {}".format(type(iris_dataset['data'])))
print("Shape of data: {}".format(iris_dataset['data'].shape))
```

## Giá trị 5 mẫu dữ liệu đầu tiên: 

```python
print("First 5 columns of data:\n{}".format(iris_dataset['data'][:5]))
```

## 'Target' là mảng 1 chiều, ma trận 150x1. Là nhãn của từng mẫu trong tập dữ liệu -> target_names

```python
print("Shape of target: {}".format(iris_dataset['target'].shape))
print("Target:\n{}".format(iris_dataset['target']))
```

## Tranining and Testing data
Sử dụng hàm *train_test_split* trong *scikit-learn* đề chia data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

X_train chiếm 75% dataset, X_test chiếm 25%

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

**alpha: độ đậm nhạt của marker <br/>
hist_kwds={'bins':20} : histogram có 20 bins <br/>
s=60: size marker <br/>
figsize: size biểu đồ <br/>**

```python
import mglearn as mglearn
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(20,20), marker='o',
                       hist_kwds={'bins':20},s=60,alpha=0.8, cmap=mglearn.cm3)
```

## First model k-Nearest Neighbors

*n_neighbors =1 : với mỗi điểm test data ta xét 1 điểm training data gần nhaast*

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)

knn.fit(X_train, y_train)
```

** *Making prediction for new data point:* <br/>**
-sepal length: 5 cm <br/>
-sepal width: 2.9 cm <br/>
-petal length: 1 cm <br/>
-petal width: 0.2 cm

```python
X_new = np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape: {}".format(X_new.shape))
```

** *Predict new data point* **

```python
prediction = knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Prediction target name: {}".format(
    iris_dataset['target_names'][prediction]))
```

## Evaluating the Model - Đánh giá Model

```python
y_pred = knn.predict(X_test)
print("Test set prediction:\n {}".format(y_pred))
```

```python
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

**Dùng *score* trong knn để tính accuracy**

```python
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```

### Whole training and evaluation procedure

```python
X_train, X_test, y_train, y_test = train_test_split(
iris_dataset['data'], iris_dataset['target'], random_state=0)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
